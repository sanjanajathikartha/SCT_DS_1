```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import zipfile
with zipfile.ZipFile("API_SP.POP.TOTL_DS2_en_csv_v2_127039 (1).zip", 'r') as zip_ref:
    zip_ref.extractall("population_data")
df = pd.read_csv(
    "population_data/API_SP.POP.TOTL_DS2_en_csv_v2_127039.csv",
    skiprows=4
)
population_2022 = df[['Country Name', '2022']]
population_2022 = population_2022.dropna()
top10 = population_2022.sort_values(
    by='2022',
    ascending=False
).head(10)
sns.set_style("whitegrid")
plt.figure(figsize=(12,6))

sns.barplot(
    x='Country Name',
    y='2022',
    data=top10
)

plt.title("Top 10 Countries by Population (2022)")
plt.xlabel("Country")
plt.ylabel("Population")

plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.hist(population_2022['2022'], bins=20)

plt.title("Distribution of Population")
plt.xlabel("Population")
plt.ylabel("Frequency")

plt.show()
```
